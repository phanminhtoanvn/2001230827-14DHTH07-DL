```python
#Bài 12
import tensorflow as tf

# 1. Viết hàm tạo 2 tensor ngẫu nhiên có kích thước (3, 2, 2)
def create_tensors():
    # Sử dụng phân phối chuẩn (normal) hoặc đều (uniform) đều được
    x = tf.random.normal(shape=(3, 2, 2))
    y = tf.random.normal(shape=(3, 2, 2))
    return x, y

# 2. Viết các hàm tính cộng, trừ, nhân, chia (Python thuần túy)
def add_func(x, y):
    return tf.add(x, y)

def sub_func(x, y):
    return tf.subtract(x, y)

def mul_func(x, y):
    return tf.multiply(x, y)

def div_func(x, y):
    return tf.divide(x, y)

# 3. Chuyển các hàm này vào trong Graph

graph_add = tf.function(add_func)
graph_sub = tf.function(sub_func)
graph_mul = tf.function(mul_func)
graph_div = tf.function(div_func)


# Gọi hàm tạo tensor
x_val, y_val = create_tensors()

# Gọi các hàm đã được chuyển sang Graph
res_add = graph_add(x_val, y_val)
res_sub = graph_sub(x_val, y_val)
res_mul = graph_mul(x_val, y_val)
res_div = graph_div(x_val, y_val)

print("Kích thước tensor x:", x_val.shape)
print("Kích thước kết quả phép CỘNG trên Graph:", res_add.shape)
print("Kích thước kết quả phép TRỪ trên Graph:", res_sub.shape)
print("Kích thước kết quả phép NHÂN trên Graph:", res_mul.shape)
print("Kích thước kết quả phép CHIA trên Graph:", res_div.shape)
```

```python
# Bài 13: Model tính tổng
class SumModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return tf.add(x, y)

# Bài 14: Model tính hiệu và tích
class SubMulModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return tf.subtract(x, y), tf.multiply(x, y)

# Chạy thử
x_val, y_val = tf.constant(10.0), tf.constant(5.0)
print("x: ", x_val.numpy(), "| y: ", y_val.numpy())
model_13 = SumModel()
print("Bài 13 - Tổng:", model_13(x_val, y_val).numpy())

model_14 = SubMulModel()
sub_res, mul_res = model_14(x_val, y_val)
print("Bài 14 - Hiệu:", sub_res.numpy(), "| Tích:", mul_res.numpy())
```

```python
# Bài 15: Tính chu vi hình chữ nhật
class RectanglePerimeter(tf.Module):
    @tf.function
    def __call__(self, a, b):
        return (a + b) * 2.0

# Bài 16: Tính diện tích toàn phần hình hộp chữ nhật
class BoxSurfaceArea(tf.Module):
    @tf.function
    def __call__(self, a, b, c):
        # S_xung_quanh = (a + b) * c * 2
        s_xq = (a + b) * c * 2.0
        # S_toan_phan = (a * b) * 2 + S_xung_quanh
        s_tp = (a * b) * 2.0 + s_xq
        return s_tp

model_15 = RectanglePerimeter()
print("Bài 15 - Chu vi HCN (a=3, b=4):", model_15(3.0, 4.0).numpy())

model_16 = BoxSurfaceArea()
print("Bài 16 - Diện tích TP hộp (a=2, b=3, c=4):", model_16(2.0, 3.0, 4.0).numpy())
```

```python
# Bài 17: Điều kiện cạnh bên song song
class ParallelSidesModel(tf.Module):
    @tf.function
    def __call__(self, AD, BC, AB, CD):
        cond1 = tf.equal(AD, BC)
        cond2 = tf.equal(AB, CD)
        return tf.logical_and(cond1, cond2)

# Bài 18: Điều kiện cạnh đáy bằng nhau
class EqualBasesModel(tf.Module):
    @tf.function
    def __call__(self, AD, BC, is_AD_parallel_BC):
        # is_AD_parallel_BC là boolean tensor biểu diễn tính song song
        cond1 = tf.equal(AD, BC)
        return tf.logical_and(cond1, is_AD_parallel_BC)

# Bài 19: Đường trung bình tam giác
class TriangleMidlineModel(tf.Module):
    @tf.function
    def __call__(self, AM, MB, AN, NC):
        # Hệ quả của đường trung bình: Nếu AM = MB và AN = NC thì M, N là trung điểm
        # Code này sẽ trả về True nếu thực sự M và N là trung điểm các cạnh
        is_M_midpoint = tf.equal(AM, MB)
        is_N_midpoint = tf.equal(AN, NC)
        return tf.logical_and(is_M_midpoint, is_N_midpoint)

model_17 = ParallelSidesModel()
print("Bài 17 - Cạnh bên song song?:", model_17(5.0, 5.0, 4.0, 4.0).numpy())

model_18 = EqualBasesModel()
print("Bài 18 - Cạnh đáy bằng nhau?:", model_18(5.0, 5.0, tf.constant(True)).numpy())

model_19 = TriangleMidlineModel()
print("Bài 19 - NA == NC (N là trung điểm)?:", model_19(3.0, 3.0, 4.0, 4.0).numpy())
```
