```python
import numpy as np

print("so sanh mang 1 chieu")
a1 = np.array([10,20,30,40])
b1 = np.array([15,20,25,50])
print("mang a1: ",a1)
print("Mang b1: ",b1)
# lon hon
print("\na1 > b1 (greater):")
print(np.greater(a1,b1))

print("\na1 >= b1 (greater_equal):")
print(np.greater_equal(a1,b1))
#nho hon
print("\na1 < b1 (greater):")
print(np.less(a1,b1))

print("\na1 <= b1 (greater_equal):")
print(np.less_equal(a1,b1))

#so sanh mang 2 chieu

print("\nSo sanh mang 2 chieu")
a2 = np.array([[1,5,9],
               [8,2,6]])
b2 = np.array([[3,5,7],
               [4,9,6]])

print("Mang a2: \n",a2)
print("Mang b2: \n",b2)

print("\na2 > b2 (greater):")
print(np.greater(a2,b2))

print("\na2 >= b2 (greater_equal):")
print(np.greater_equal(a2,b2))
#nho hon
print("\na1 < b1 (greater):")
print(np.less(a1,b1))

print("\na1 <= b1 (greater_equal):")
print(np.less_equal(a1,b1))
```

```python
import numpy as np

# Tạo mảng gồm 10 số 1
array_ones = np.ones(10)

# Tạo mảng gồm 10 số 0
array_zeros = np.zeros(10)

# Tạo mảng gồm 10 số 5 (sử dụng hàm np.full để điền toàn bộ bằng số 5)
array_fives = np.full(10, 5)

# Nối 3 mảng lại với nhau thành một mảng 1 chiều duy nhất
result_array = np.concatenate((array_ones, array_zeros, array_fives))

print("Mảng kết quả:\n", result_array)
print("\nKích thước của mảng:", result_array.shape)
```

```python
x_20 = pd.Series(np.random.randint(1, 50, 20))
y_20 = pd.Series(np.random.randint(1, 50, 20))

# Hội (Union) - Lấy tất cả các phần tử duy nhất của cả 2 mảng
union_result = np.union1d(x_20, y_20)

# Giao (Intersect) - Lấy các phần tử xuất hiện ở cả 2 mảng
intersect_result = np.intersect1d(x_20, y_20)

print("Tập hợp Hội (Union):\n", union_result)
print("\nTập hợp Giao (Intersect):\n", intersect_result)
```

```python
x_21 = pd.Series([0, 1, 2, 7, 8, 9])
y_21 = pd.Series(['p', 'q', 'r', 'w', 'x', 'y'])

# Nối chiều dọc (vertical) - tạo thành Series dài hơn
vertical_concat = pd.concat([x_21, y_21], axis=0, ignore_index=True)
print("Nối dọc (Series):\n", vertical_concat)

# Nối chiều ngang (horizontal) - tạo thành DataFrame
horizontal_concat = pd.concat([x_21, y_21], axis=1)
print("\nNối ngang (DataFrame):\n", horizontal_concat)
```

```python
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df_exam = pd.DataFrame(exam_data, index=labels)
print("--- Bài 22: DataFrame ban đầu ---\n", df_exam)

# Bài 23: Cho biết 3 dòng đầu tiên
print("\n--- Bài 23: 3 dòng đầu tiên ---")
print(df_exam.head(3))

# Bài 24: Chọn cột 'name' và 'score'
print("\n--- Bài 24: Cột 'name' và 'score' ---")
print(df_exam[['name', 'score']])

# Bài 25: Chọn những dòng có attempts > 2
print("\n--- Bài 25: attempts > 2 ---")
print(df_exam[df_exam['attempts'] > 2])

# Bài 26: Cho biết số dòng, số cột của DataFrame
rows, cols = df_exam.shape
print(f"\n--- Bài 26: Số dòng, số cột ---\nSố dòng: {rows}, Số cột: {cols}")

# Bài 27: Những dòng có score là NaN
print("\n--- Bài 27: score là NaN ---")
print(df_exam[df_exam['score'].isnull()])

# Bài 28: Những dòng có score trong khoảng (15, 20) (tính cả 15 và 20)
print("\n--- Bài 28: score trong khoảng [15, 20] ---")
print(df_exam[df_exam['score'].between(15, 20)])

# Bài 29: Những dòng có attempt > 2 VÀ score trong khoảng (15, 20)
print("\n--- Bài 29: attempts > 2 VÀ score trong khoảng [15, 20] ---")
# Lưu ý đặt điều kiện trong dấu ngoặc tròn () khi dùng toán tử &
print(df_exam[(df_exam['attempts'] > 2) & (df_exam['score'].between(15, 20))])

# Bài 30: Thay đổi giá trị dòng 'd', cột score là 19 (sử dụng .loc)
print("\n--- Bài 30: Thay đổi score dòng 'd' thành 19 ---")
df_exam.loc['d', 'score'] = 19
print(df_exam)
```
